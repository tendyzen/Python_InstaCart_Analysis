# Cleaning Values
We will replace them with null values.

```python
df_ords_merged.isnull().sum()
```

```python
#replace values > 100$ with NaN values

df_ords_merged.loc[df_ords_merged['prices']>100, 'prices'] = np.nan

df_ords_merged['prices'].isnull().sum()
```

```python
df=df_ords_merged[:1000000]
df.head(), df.shape
```

```python
df.groupby('product_name')
```

Using .agg() returns a dataframe and allows a dictionary to apply multiple functions at once to the data.

```python
df.groupby('department_id').agg({'customer_order_number' : ['mean']})
```

Using this direct method returns a series, and is for only a single aggregation function 

```python
df.groupby('department_id')['customer_order_number'].mean()
```

Using .agg(), you can apply multiple functions to data using a list as a value for the key column

```python
df.groupby('department_id').agg({'customer_order_number' : ['mean', 'min', 'max']})
```

```python
df_ords_merged['max_orders'] = df_ords_merged.groupby('customer_id')['customer_order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
df_ords_merged[['customer_id', 'max_orders']].head(100)
```

# Feature Binning
Create Loyalty Flag Column

```python
df_ords_merged.loc[df_ords_merged['max_orders'] > 40, 'loyalty_flag'] = 'Loyal Customer'
df_ords_merged.loc[df_ords_merged['max_orders'] < 10, 'loyalty_flag'] = 'New Customer'
df_ords_merged.loc[(df_ords_merged['max_orders'] >= 10) & (df_ords_merged['max_orders'] <= 40), 'loyalty_flag'] = 'Regular Customer'

df_ords_merged.head(100)
```

```python
#printing frequency counts

df_ords_merged['loyalty_flag'].value_counts()
```

```python
df_ords_merged[['customer_id', 'customer_order_number', 'max_orders', 'loyalty_flag']].head(100)
```

Comparing Customer's Average Spend

```python
#create new column for avg customer spend

df_ords_merged['avg_cust_spend'] = df_ords_merged.groupby('customer_id')['prices'].transform('mean')
```

```python
df_ords_merged.head()
```

```python
#create new flag column for customer spend

df_ords_merged.loc[df_ords_merged['avg_cust_spend'] < 10, 'Customer_Spend'] = 'Low Spender'
df_ords_merged.loc[df_ords_merged['avg_cust_spend'] >= 10, 'Customer_Spend'] = 'High Spender'
```

```python
df_ords_merged['Customer_Spend'].value_counts()
```

Comparing Customer's Order Frequency

```python
#create new column with median time between orders for each customer id

df_ords_merged['median_order_frequency'] = df_ords_merged.groupby('customer_id')['days_between_last_order'].transform('median')
```

```python
df_ords_merged.loc[df_ords_merged['median_order_frequency'] > 20, 'Order_Frequency'] = 'Non-Frequent Spender'
df_ords_merged.loc[(df_ords_merged['median_order_frequency'] <= 20) & (df_ords_merged['median_order_frequency'] >= 10), 'Order_Frequency'] = 'Regular Spender'
df_ords_merged.loc[df_ords_merged['median_order_frequency'] < 10, 'Order_Frequency'] = 'Frequent Spender'
```

```python
#checking new frequency columns
df_ords_merged['Order_Frequency'].value_counts()
```
