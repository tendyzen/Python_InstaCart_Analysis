# change data types

```python
# Define the columns to cast to int16 and float16
int16_columns = ['customer_order_number', 'orders_day_of_week', 'order_hour_of_day', 
                 'add_to_cart_order', 'reordered', 'Unnamed: 0_y', 'aisle_id', 
                 'department_id', 'max_orders']
float16_columns = ['days_between_last_order', 'prices', 'avg_cust_spend', 'median_order_frequency']

# Cast the selected columns to int16 and float16
df_vis[int16_columns] = df_vis[int16_columns].astype('int16')
df_vis[float16_columns] = df_vis[float16_columns].astype('float16')

#check df_vis
df_vis.dtypes
```

```python
# Perform the merge
df = df_vis.merge(df_cust, on='customer_id')

#check new df
df.describe(), df.shape, df.isnull().sum()
```

# clean
