# feature analysis
Relation between Age and Dependents

```python
# create efficient df

df_fam = df_vis[['Age', 'Dependents']]

#sample df
df_fam_sample = df_fam.sample(frac=0.3, random_state=4)
df_fam_sample.shape
```

```python
#create line chart exploring relation between age and dependents
line2 = sns.lineplot(x='Age', y='Dependents', data=df_fam_sample)
```

```python
line3 = sns.lineplot(x='Dependents', y='Age', data=df_fam_sample)
```

```python
#save chart
line2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Age_Dependents_Line2.png'))
line3.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Age_Dependents_Line3.png'))
```

This chart explores a relationship between age of customer and the number of dependents. 
It shows that customers are more likely to have a dependent in middle ages.
