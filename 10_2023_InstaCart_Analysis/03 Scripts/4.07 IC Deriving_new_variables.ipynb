# Feature binning
Define labeling function

```python
#define function to make price categories
def price_label(row):
    if row['prices'] < 5:
        return 'Low Range Item'
    elif row['prices'] >= 5 and row['prices'] < 15:
        return 'Mid Range Item'
    elif row['prices'] >= 15:
        return 'High Range Item'
    else: return 'Not Enough Data'
```

```python
#apply function to shortened dataframe: df
df['Price Labels'] = df.apply(price_label, axis=1)
df.head()
```

```python
#count the results of the labeling function
df['Price Labels'].value_counts()
```

```python
#check max price in the df subset
df['prices'].max()
```

Use Loc instead of a function for labeling prices

```python
#labeling low prices
df_ords_merged.loc[df_ords_merged['prices'] <= 5, 'price_range_loc'] = 'Low Range Items'
```

```python
#labeling mid prices
df_ords_merged.loc[(df_ords_merged['prices'] > 5) & (df_ords_merged['prices'] <= 15), 'price_range_loc'] = 'Mid Range Items'
```

```python
#labeling high prices
df_ords_merged.loc[df_ords_merged['prices'] > 15, 'price_range_loc'] = 'High Range Items'
```

```python
#showing the results of labeling with loc
df_ords_merged['price_range_loc'].value_counts()
```

Inspecting traffic throughout the week

```python
#count orders for days of week
df_ords_merged['orders_day_of_week'].value_counts()
```

labeling most and least busy days in new columns

```python
#create empty list
result = []

#fill list using for loop
for value in df_ords_merged['orders_day_of_week']:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append("Least Busy Day")
    else:
        result.append("Regularly Busy Day")
        
```

```python
#add result list to df
df_ords_merged['Busiest Day'] = result
df_ords_merged['Busiest Day'].value_counts()
```

Make a new column for the Two Most and Two Least busy days

```python
#labeling two most and two least busy days in new columns. most: 0 and 1. least: 4 and 3.

#create empty list
results = []

#fill list using for loop
for value in df_ords_merged['orders_day_of_week']:
    if value == 0:
        results.append("Busiest Day")
    elif value == 1:
        results.append("Busiest Day")
    elif value == 4:
        results.append("Least Busy Day")
    elif value == 3:
        results.append("Least Busy Day")
    else:
        results.append("Regularly Busy Day")
results[:20]
```

```python
#add results as a new column to df_ords_merged df

df_ords_merged['Busiest Days'] = results
df_ords_merged.head()
```

```python
df_ords_merged['Busiest Days'].value_counts(), df_ords_merged['orders_day_of_week'].value_counts()
```

This seems all correct, as days 0, 1, 3, and 4 are represented as top 2 least and most busy days. days 6, 2, and 5 are combined as "regularly busy days".

Create a new column containing these labels called “busiest_period_of_day.”

```python
#value count the hours of day column
df_ords_merged['order_hour_of_day'].value_counts(), df_ords_merged['order_hour_of_day'].describe()
```

```python
#identify the busiest hours of the day. Rather than by hour, they want periods of 
#time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
```

Most Orders: 9, 10, 11, 12, 13, 14, 15, 16 (Top 1/3 of hours)
Average Orders: 7, 8, 17, 18, 19, 20, 21, 22 (Middle 1/3 of hours)
Fewest Orders: 23, 0, 1, 2, 3, 4, 5, 6 (Bottom 1/3 of hours)

```python
#create lists for the categories of hours, and the list / for loop that will become the column, busiest_period_of_day
Most_Orders= [9, 10, 11, 12, 13, 14, 15, 16]
Average_Orders= [7, 8, 17, 18, 19, 20, 21, 22]
Fewest_Orders= [23, 0, 1, 2, 3, 4, 5, 6]
    
list = []

for value in df_ords_merged['order_hour_of_day']:
    if value in Most_Orders:
        list.append("Most_Orders")
    elif value in Average_Orders:
        list.append("Average_Orders")
    elif value in Fewest_Orders:
        list.append("Fewest_Orders")
    else:
        list.append("No Data")
        
```

```python
#checking list
list[:10]
```

```python
#add new column to df
df_ords_merged['busiest_hours_of_day'] = list
df_ords_merged.head()
```

```python
#checking for consistency
df_ords_merged['busiest_hours_of_day'].value_counts()
```

```python
#drop busiest day column and check
df_ords_merged = df_ords_merged.drop('Busiest Day', axis=1)
```
